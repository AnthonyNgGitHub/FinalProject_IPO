## Grabbing all filtered cash_flow CSV's to put into 1 file.

```python
# Importing dependencies.
from glob import glob
import pandas as pd
```

```python
# Creating a list of all the balance_sheet files.
cash_flow_files = sorted(glob('raw_dataset/filtered_cash_flow/*filtered_cash_flow.csv'))
cash_flow_files
```

```python
# Create a dataframe to house all concatenated files.
df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in cash_flow_files), ignore_index = True)
df
```

```python
# Dropping unwanted columns.
cleaned_df = df.drop(['filename'],axis=1)
cleaned_df
```

```python
# Count of missing values of the entire dataframe. If any drop.
cleaned_df.isnull().sum().sum()
```

```python
# Replacing the None values with 0.
filtered_cash_flow_df = cleaned_df.replace('None', 0)
filtered_cash_flow_df
```

```python
# Renaming column names.

filtered_cash_flow_df = filtered_cash_flow_df.rename(columns = {
    'SYMBOL': 'symbol', 
    'fiscalDateEnding': 'fiscaldate_end', 
    'reportedCurrency': 'reported_currency',
    'operatingCashflow': 'operating_cashflow',
    'paymentsForOperatingActivities': 'payments_foroperating_activities', 
    'proceedsFromOperatingActivities': 'proceeds_fromoperating_activities', 
    'changeInOperatingLiabilities': 'changein_operating_liabilities',
    'changeInOperatingAssets': 'changein_operating_assets',
    'depreciationDepletionAndAmortization': 'dpr_depletion_amort', 
    'capitalExpenditures': 'capital_expenditures', 
    'changeInReceivables': 'changein_receivables',
    'changeInInventory': 'changein_inventory',
    'profitLoss': 'profit_loss', 
    'cashflowFromInvestment': 'cashflow_frominvestment', 
    'cashflowFromFinancing': 'cashflow_fromfinancing',
    'proceedsFromRepaymentsOfShortTermDebt': 'proceeds_fromrepayment_shortterm_debt',
    'paymentsForRepurchaseOfCommonStock': 'payments_forrepurchase_commonstock', 
    'paymentsForRepurchaseOfEquity': 'payments_forrepurchase_equity', 
    'paymentsForRepurchaseOfPreferredStock': 'payments_forrepurchase_preferredstock',
    'dividendPayout': 'div_payout',
    'dividendPayoutCommonStock': 'div_payout_commonstock', 
    'dividendPayoutPreferredStock': 'div_payout_preferredstock', 
    'proceedsFromIssuanceOfLongTermDebtAndCapitalSecuritiesNet': 'proceeds_fromissuance_longterm_debtcapital_secnet',
    'proceedsFromIssuanceOfPreferredStock': 'proceeds_fromissuance_preferredstock',
    'proceedsFromRepurchaseOfEquity': 'proceeds_fromrepurchase_equity', 
    'proceedsFromSaleOfTreasuryStock': 'proceeds_fromsale_treasurystock',
    'changeInCashAndCashEquivalents': 'changein_cash_cashequivalents',
    'changeInExchangeRate': 'changein_exchangerate',
    'netIncome': 'netincome',
    })
filtered_cash_flow_df
```

```python
# Exporting balance_sheet dataframe to csv file without the index.
filtered_cash_flow_df.to_csv('raw_dataset/filtered_cash_flow.csv', index = False)
```
