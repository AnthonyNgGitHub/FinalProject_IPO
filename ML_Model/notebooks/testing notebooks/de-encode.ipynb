```python
import pandas as pd
```

# First Day

```python
first_day_df = pd.read_csv('../first_day_results.csv')
first_day_df
```

```python
def undummify(df, prefix_sep="_"):
    cols2collapse = {
        item.split(prefix_sep)[0]: (prefix_sep in item) for item in df.columns
    }
    series_list = []
    for col, needs_to_collapse in cols2collapse.items():
        if needs_to_collapse:
            undummified = (
                df.filter(like=col)
                .idxmax(axis=1)
                .apply(lambda x: x.split(prefix_sep, maxsplit=1)[1])
                .rename(col)
            )
            series_list.append(undummified)
        else:
            series_list.append(df[col])
    undummified_df = pd.concat(series_list, axis=1)
    return undummified_df
```

```python
undummied_df = undummify(first_day_df)
```

```python
undummied_df
```

```python
new_first_day_df = first_day_df[['debt_asset_ratio','debt_equity_ratio','current_ratio','OCF_ratio','gross_profit_margin','net_profit_margin','cashflow_frominvestment','EBIT_margin','actual_first_day_gain_loss','predicted_range','predicted_first_day_gain_loss']]
new_first_day_df
```

```python
new_first_day_df['sector'] = undummied_df['sector']
new_first_day_df
```

```python
new_first_day_df['industry'] = undummied_df['industry']
new_first_day_df
```

```python
new_first_day_df.to_csv('first_day_results_de-encoded.csv', index=False)
```
