##### To filter the balance sheet of each stock for the record closest to the IPO date.

```python
import pandas as pd
import datetime
from datetime import timedelta
```

```python
d_parser = lambda x: pd.datetime.strptime(x,'%Y-%m-%d')
IPO_df = pd.read_csv('../../raw_dataset/IPO_SCOOP_Listing.csv', parse_dates = ['Trade Date'], date_parser = d_parser)
IPO_df
```

```python
IPO_df.dtypes
```

```python
IPO_list = IPO_df['Symbol'].tolist()
IPO_list
```

```python
# Check length of IPO_list
len(IPO_list)
```

#### Create a loop to perform a filter on each statement for each stock

```python
# Loop through all symbols in IPO_List

FileNotFoundError_List = []

for symbol in IPO_list:
    
    
    try: 

        symbol_df = IPO_df.loc[IPO_df['Symbol']==symbol]
        symbol_IPO_DATE = symbol_df['Trade Date']

        # Define range lower bound and upper bound
        delta_days = 45

        date_lower_bound = symbol_IPO_DATE - timedelta(days=delta_days)
        date_upper_bound = symbol_IPO_DATE + timedelta(days=delta_days)

        print(f'{symbol}: {symbol_IPO_DATE}')
        print(f'{symbol} lower bound: {date_lower_bound}')
        print(f'{symbol} upper bound: {date_upper_bound}')
        print(f'----------------------------------')

        # Define dynamic path for each symbol
        symbol_balance_sheet_path = f'../../raw_dataset/balance_sheet/{symbol}_balance_sheet.csv'

        # Read in each symbol's income statement
        d_parser = lambda x: pd.datetime.strptime(x,'%Y-%m-%d')
        symbol_balance_sheet_df = pd.read_csv(symbol_balance_sheet_path, parse_dates = ['fiscalDateEnding'], date_parser = d_parser, index_col=None)
        pd.set_option('display.max_columns',85)

        # To filter for dates within the IPO_date +/- 30 days range
        fiscalDate_filt = (symbol_balance_sheet_df['fiscalDateEnding'] >= date_lower_bound.squeeze()) & (symbol_balance_sheet_df['fiscalDateEnding'] <= date_upper_bound.squeeze())
        filtered_symbol_balance_sheet_df = symbol_balance_sheet_df.loc[fiscalDate_filt]

        # Exporting test income statement dataframe to csv file
        save_file_path = f'../../raw_dataset/filtered_balance_sheet/{symbol}_filtered_balance_sheet.csv'
        filtered_symbol_balance_sheet_df.to_csv(save_file_path, index=False)
        
    except FileNotFoundError:
        print(f"FileNotFoundError occured on {symbol}")
        print(f'----------------------------------')
        FileNotFoundError_List.append(symbol)
```
